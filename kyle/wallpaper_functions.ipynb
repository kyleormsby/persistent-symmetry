In this notebook I'll follow Farriss-Lanning (https://core.ac.uk/download/pdf/82036994.pdf) to produce smooth functions with perscribed symmetry. (Note the typo in the formul√¶ for $X,Y$ on p.198: $2\pi$ should be in the numerator, not the denominator.)

The following function has $p6$ symmetry:

```python
var('x y')
Xhex = 2*pi*(x+y/sqrt(3))
Yhex = 2*pi*y/sqrt(3)

fp6(x,y) = cos(Xhex+Yhex)+cos(Xhex-2*Yhex)+cos(-2*Xhex+Yhex)
win=5
contour_plot(fp6,(-win,win),(-win,win),cmap='viridis',plot_points=500,colorbar=False,frame=False,contours=20)
```

Here's one with $p3$:

```python
fp3 = sin(Xhex+Yhex)+sin(Xhex-2*Yhex)+sin(-2*Xhex+Yhex)
contour_plot(fp3,(-5,5),(-5,5),cmap='viridis',plot_points=500,colorbar=False,frame=False,contours=20)
```

For the square lattice and $p4$ symmetry, we can use standard $xy$-coordinates (though note that $X=y, Y=x$ in the paper's notation).

```python
fp4(x,y) = cos(x)*cos(y)
contour_plot(fp4,(-10,10),(-10,10),cmap='viridis',plot_points=500,colorbar=False,frame=False,contours=20)
```

Now I'll attempt $p1$, but at least on visual inspection this seems to give $p2$ symmetry ‚Äî comments welcome. We'll use the lattice generated by $\tau_1 = (1,0)$ and $\tau_2 = (1/2,1)$.

```python
Xgen = 2*pi*(x-y/2)
Ygen = 2*pi*y

fp1(x,y) = cos(3*Xgen+2*Ygen)+sin(3*Xgen+2*Ygen)+cos(2*Xgen+Ygen)+sin(2*Xgen+Ygen) #+cos(4*Xgen+Ygen)+sin(4*Xgen+Ygen)+cos(5*Xgen+2*Ygen)+sin(5*Xgen+2*Ygen)
win = 5
contour_plot(fp1,(-win,win),(-win,win),cmap='viridis',plot_points=500,colorbar=False,frame=False,contours=20)
```

Finally, we produce a smooth function with $p2$ symmetry. This uses the rectangular lattice generated by $\tau_1 = (2,0)$ and $\tau_2 = (0,1)$.

```python
Xrec = pi*x
Yrec = 2*pi*y

# Xrec = 2*pi*(x-y/2)
# Yrec = 2*pi*y

fp2(x,y) = cos(Xrec)*cos(Yrec)+cos(2*Xrec)*cos(Yrec)+cos(5*Xrec)*cos(3*Yrec)

win = 3
contour_plot(fp2,(-win,win),(-win,win),cmap='viridis',plot_points=500,colorbar=False,frame=False,contours=20)
```
