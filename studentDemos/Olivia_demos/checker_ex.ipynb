```python
from ripser import ripser
from persim import plot_diagrams
import numpy as np
from numpy import pi,cos,sin
import matplotlib.pyplot as plt
import random
```

## Define function on all of $\mathbb{R}^2$ that will look like a checker board

```python
# 0 corresponds to white, 1 corresponds to black
def checker(x,y): 
    if 0 <= x%2 < 1 and 0 <= y%2 < 1:
        res = 1 
    elif 1 <= x%2 < 2 and 1 <= y%2 < 2:
        res = 1
    else: 
        res = 0
    return res
```

### Taking a sample of points from $\mathbb{R}^2$ to visualize the checker board (this part is not necessary)

```python
n_samples = 100000
X = list(20 * np.random.random(n_samples) - 10) # x-value from (-10,10)
Y = list(20 * np.random.random(n_samples) - 10) # y-value from (-10,10)

points = []
for i in range(n_samples):
    color = checker(X[i],Y[i])
    if color == 1:
        points.append((X[i],Y[i])) # only include the points where the checker function outputs 1
        
x = []
y = []
for i in points: 
    x.append(i[0])
    y.append(i[1])
```

```python
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes()
 
# Creating plot
ax.scatter(x, y, s = .1)
plt.title("Checker Board")
 
# show plot
plt.show()
```

## Making Lens

```python
# take sample from (-2, 2) at specfied increments (i.e. not a random sample)
# thus our lens/window is (-2,2) x (-2,2)
num_pixels = 40 # this is really the square root of the total number of pixels
sample = list(np.linspace(-1.95,1.95,num_pixels))
```

```python
# once we have a sample, define pixels at the cartesian product of this sample 

def get_pixels(sample):
    window_sample = []
    for i in range(len(sample)):
        for j in range(len(sample)):
            window_sample.append((sample[i],sample[j]))
    return window_sample

pixels = get_pixels(sample)
pixels
```

```python
# apply the checker function (color_function in this case) to get the color associated with each pixel

def get_colors(pixels, color_function):
    colors = []
    for pixel in pixels:
        colors.append(color_function(pixel[0],pixel[1]))
    return colors   

colors = get_colors(pixels,checker)
colors
```

## Defining isometries

```python
def translation(a,b):
    def t_ab(x,y): 
        return x+a, y+b
    return t_ab
```

```python
def rotation(theta):
    def r_theta(x,y):
        return x * cos(theta) - y * sin(theta), x * sin(theta) + y * cos(theta)
    return r_theta
```

```python
def isom(a,b,theta):
    def iso(x,y):
        r = rotation(theta)
        t = translation(a,b)
        return t(r(x,y)[0], r(x,y)[1])
    return iso
```

```python
# isometries can be parameterized by a point in R^2 and an angle 
# this requires three pararmeters, which I am denoting a, b, theta
# here I randomly sample 1000 values for each parameter a, b, theta

n_samples = 1000
a = list(200 * np.random.random(n_samples) - 100) # sampling from (-100,100)
b = list(200 * np.random.random(n_samples) - 100) # sampling from (-100,100)
theta = list(2 * pi * np.random.random(n_samples)) # sampling from (0, 2pi)

# define isometries (and their inverses) based on these parameters
isos = []
isos_inv = []
for i in range(1000):
    isos.append(isom(a[i], b[i], theta[i]))
    isos_inv.append(isom(-1 * a[i],-1 * b[i],2 * pi - theta[i])) 
```

### Let's take a closer look at one of these isometries I've just generated ...

```python
# for each pixel (x,y), pixel_inv is f_inv(x,y), 
# under the isometry f, the new color of pixel (x,y) will be the original color of the pixel f_inv(x,y)

pixels_inv = []
for pixel in pixels: 
    pixels_inv.append(f_inv(pixel[0], pixel[1]))
new_colors = get_colors(pixels_inv, checker)

# this describes the same set of pixels (our lens/window) and their new coloring
for i in range(len(pixels_inv)):
    print ((pixels[i],new_colors[i]))
```

## Generating representative vectors in $\mathbb{R}^{1600}$ for each of these isometries

```python
# same procedure as above, but now a generalizable function to get the vector associated to each isometry
def new_colors(iso_inv):
    pixels_inv = []
    for pixel in pixels: 
        pixels_inv.append(iso_inv(pixel[0], pixel[1]))
    new_colors = get_colors(pixels_inv, checker)
    return new_colors

# apply this function to each (inverse) isometry 
vectors = []
for i in isos_inv:
    vectors.append([new_colors(i)])
```

## Calculating the persistent homology of the orbit space!

```python
M = np.concatenate(vectors, axis=0 ) # ripser takes an array as arguement, so we concatenate the vectors
M.shape # we have 1000 lens/windows (each associated to a different isometry) made up of 1600 pixels
```

```python
# coefficients in Z2
# this will take a couple minutes to run!
output2 = ripser(M, maxdim=1, coeff=2)
dgms2 = output2['dgms']
fig = plt.figure(figsize = (12, 6))
plot_diagrams(dgms2)
```
